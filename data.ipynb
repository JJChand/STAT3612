```python
import numpy as np
import pandas as pd
import json
import os

test_patient_info = pd.read_csv('kaggle-dataset/clinical_information/test_patient_info.csv')
train_patient_info = pd.read_csv('kaggle-dataset/clinical_information/train_patient_info.csv')
val_patient_info = pd.read_csv('kaggle-dataset/clinical_information/val_patient_info.csv')

# Load JSON files correctly
with open('kaggle-dataset/train.json', 'r') as f:
    train_json = json.load(f)

with open('kaggle-dataset/test.json', 'r') as f:
    test_json = json.load(f)

with open('kaggle-dataset/val.json', 'r') as f:
    val_json = json.load(f)

# Convert nested JSON to DataFrame
rows = []

for case_id, info in train_json.items():
    row = info.copy()
    row["case_id"] = case_id
    rows.append(row)
    
train_json1 = pd.DataFrame(rows)

train_json_clean = train_json1[["case_id","Overall_class"]] #our clean y 

for case_id, info in val_json.items():
    row = info.copy()
    row["case_id"] = case_id
    rows.append(row)

val_json1 = pd.DataFrame(rows)

val_json_clean = val_json1[["case_id","Overall_class"]]
```
